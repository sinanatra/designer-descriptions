# Image Umap

This notebook loads a csv and builds a Umap visualisation around one metadata column

### Import Variables

```python
import matplotlib.pyplot as plt
import cv2
import os
import random
import numpy as np
from PIL import Image
from sklearn.feature_extraction import image
from sklearn.manifold import TSNE
from tqdm import tqdm
```

### Parse all images in custom folder, saves their color and path into the data array

```python
data = []
folder = 'img/'

for filename in tqdm((os.listdir(folder))):
    image = cv2.imread(os.path.join(folder,filename))
    if image is not None:
        # image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image = cv2.resize(image, (32,32))
        image = image.flatten()
        data.append([image, folder + filename])
```

### Training

```python
features, images  = zip(*data)
```

```python
from sklearn.decomposition import PCA

features = np.array(features)
pca = PCA(n_components = 132)
pca.fit(features)
pca_features = pca.transform(features)
```

```python
import umap.umap_ as umap
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

reducer = umap.UMAP(
        n_neighbors = 3,
        min_dist = 1
)

X = np.array(pca_features)
# pipe = Pipeline([('scaling', StandardScaler()), ('umap', reducer)])

pipe = TSNE(n_components=2, learning_rate=500, perplexity=100, angle=0.2, verbose=0)


embedding = pipe.fit_transform(X)

tx, ty = embedding[:,0], embedding[:,1]
tx = (tx-np.min(tx)) / (np.max(tx) - np.min(tx))
ty = (ty-np.min(ty)) / (np.max(ty) - np.min(ty))
import matplotlib.pyplot
from matplotlib.pyplot import imshow

width = 4000
height = 3000
max_dim = 32

full_image = Image.new('RGBA', (width, height))
for img, x, y in zip(images, tx, ty):
    tile = Image.open(img)
    rs = max(1, tile.width/max_dim, tile.height/max_dim)
    tile = tile.resize((int(tile.width/rs), int(tile.height/rs)), Image.LANCZOS)
    full_image.paste(tile, (int((width-max_dim)*x), int((height-max_dim)*y)), mask=tile.convert('RGBA'))

matplotlib.pyplot.figure(figsize = (16,12))
imshow(full_image)
```

### Save a json and a png of the cluster

```python
import os
import json

full_image.save("/data/umap.png")
umap_path = "/data/umap.json"

data = [{"path":img, "point":[float(x), float(y)]} for img, x, y in zip(images, tx, ty)]
with open(umap_path, 'w') as outfile:
    json.dump(data, outfile)

print("saved UMAP result to %s" % umap_path)
```
